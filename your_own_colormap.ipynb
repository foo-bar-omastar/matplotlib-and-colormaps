```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import math
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
```

```python
#Loading the Matlab txt file into python notebook
cmap_vector = np.loadtxt("S:/00 MPENM/YEAR 1/Study Period 2/Scientific Visualization/20191115 - Week Ending/CubeHelix-master/CubeHelix-master/mycolormap.txt")
```

```python
#Printing the loaded vector for reference purposes
#cubehelix(128,2.64,0.61,1.74,1.47,[0.03,0.91],[0.02,1.00])
print(cmap_vector)
```

```python
#Getting the dimensions of the vector
cmap_vector.shape
```

```python
#Adding a 4th column of 1's to the RGB array to make it an RGBA array as numpy takes in a Nx4 array for the listed color map

#Creating a column vector with teh same number of rows as the RGB vector
vals = np.ones((128, 1))
#Appending the column to the end of the RGB Vector
cmap_vector = np.append(cmap_vector, vals, axis=1)
```

```python
#Printing the RGBA Vector
print(cmap_vector)
```

```python
#Creating the Listed colormap with the consolidated information from above
mycmap = ListedColormap(cmap_vector)
```

```python
#Using linspace to create the plot used in the previous exercise
def f(x,y):
    return y * np.cos(x/2)

#Defining the range of calues x and y need to take, along with the length in order to correctly set the dimension of the image
x = np.linspace(-np.pi, np.pi,1024)
y = np.linspace(0., 1.,512)
X,Y = meshgrid(x, y) # grid of point

Z = f(X, Y) # evaluation of the function on the grid

plt.figure(figsize=(15, 5))


plt.subplot(121)
plt.title('Own Colormap')

#Plotting the function using the custom created colormap
plt.imshow(Z, cmap=mycmap)
plt.colorbar(imshow(Z, cmap=mycmap))

plt.subplot(122)
plt.title('Grayscale')

plt.imshow(Z, cmap=cm.gray)
plt.colorbar(imshow(Z, cmap=cm.gray))

plt.suptitle('Own Colormap vs Grayscale Comparison')
plt.show()
```
