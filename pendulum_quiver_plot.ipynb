```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap, LinearSegmentedColormap, Normalize

#Importing the Custom Colormap I Created Earlier
cmap_vector = np.loadtxt("S:/CubeHelix-master/CubeHelix-master/mycolormap.txt")
vals = np.ones((128, 1))
cmap_vector = np.append(cmap_vector, vals, axis=1)
mycmap = ListedColormap(cmap_vector)
```

```python

x = np.arange(-np.pi,np.pi,0.5)
y = np.arange(-3,3,0.5)

#Generating a mesh of points based on the initial values
X, Y = np.meshgrid(x, y)
u = Y
v = -np.sin(X)

#Since the quiver() function has a 5th Argument '[C]' that can take in an array as input for color, we define it to be the
#formula for the magnitude of the vectors. Using the optional cmap argument, we pass our custom made colormap accordingly
fig, ax = plt.subplots()
ax.quiver(X,Y,u,v,np.sqrt(u**2 + v**2), cmap=mycmap)

#Creating a colorbar for the quiverplot
norm = Normalize()
norm.autoscale(cmap_vector)
cm = mycmap
sm = plt.cm.ScalarMappable(cmap=cm, norm=norm)
sm.set_array([])
plt.colorbar(sm)

#Displaying the Quiver plot
plt.title('Pendulum Quiver Plot')
plt.show()
```
