```python
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
```

```python
#Defining the function to be calculated
def f(x,y):
    return y * np.cos(x/2)

#Defining the range of values x and y need to take, along with the length in order to correctly set the dimension of the image
x = np.arange(-np.pi, np.pi,(2*np.pi)/1024)
y = np.arange(0., 1., 1/512)
X,Y = meshgrid(x, y) # grid of point

Z = f(X, Y) # evaluation of the function on the grid

im = imshow(Z, cmap=cm.gray) # drawing the function
colorbar(im) # adding the colobar on the right
# latex fashion title
title('$f(x,y) = y\cdot{cos(x/2)}$')

show()

plt.imsave('saved_img.jpg',Z, cmap=cm.gray )
```

```python
#Using linspace
def f(x,y):
    return y * np.cos(x/2)

#Defining the range of calues x and y need to take, along with the length in order to correctly set the dimension of the image
x = np.linspace(-np.pi, np.pi,1024)
y = np.linspace(0., 1.,512)
X,Y = meshgrid(x, y) # grid of point

Z = f(X, Y) # evaluation of the function on the grid

im = imshow(Z, cmap=cm.gray) # drawing the function
colorbar(im) # adding the colobar on the right
# latex fashion title
title('$f(x,y) = y\cdot{cos(x/2)}$')

show()
```
