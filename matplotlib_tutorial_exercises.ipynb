```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.ylabel('y axis label')
plt.show()
#Here we see that if only one argument is specified to the plot() function, then it assumes those values to be for
#the y axis by default. Hence the values 1,2,3,4 are on the y axis. Based on those values, it automatically generates 
#x axis values which we see to be 0,1,2,3 in this case
```

```python
#This command basically plts x versus y graphs as there are 2 arguments passed to the plot() function.
#Be sure to specify the same number of entries for each of the axis to avoid any error. 
plt.plot([2,3,7,9,10],[1,2,3,4,5])
plt.show()
```

```python
#Now we will try to format the stle of the plot.
#The plot() function actually takes in a third optional argument that specifies the line color and style of the plot
#These two aspects of the plot format can be passed as a concatenated string in the plot() function call
#By default the style is a 'b-' where 'b' = blue and '-' = line

#Let us try to format the following plot with red dote by specifying 'ro' as the optional argument
plt.plot([1,2,3,4,5,6],[3,1,5,4,7,9],'ro')
plt.axis([0,7,0,10])
plt.show()
```

```python
#As we see above, we have added an additional command '.axis()' which takes a list of values to
#specify the range of min and max values for the x and y axis respectively
#It is of the form [xmin, xmax, ymin, ymax]
```

```python
#Now let us look up more line specifications from matlab : https://se.mathworks.com/help/matlab/ref/linespec.html
#We use the same example as above but this time we try out a couple of different specifiers to see how they work
```

```python
#Format Type 1 - Dot Dash | Magenta | Asterisk
plt.plot([1,2,3,4,5,6],[3,1,5,4,7,9],'-.m*')
plt.axis([0,7,0,10])
plt.show()
```

```python
#Format Type 2 - Dotted | Cyan | Hexagram
plt.plot([1,2,3,4,5,6],[3,1,5,4,7,9],':ch')
plt.axis([0,7,0,10])
plt.show()
```

```python
#Now let us plot graphs with keyword strings
#Instances where we have data stored in arrays, or in variables in general and the entire set of data needs to be called,
#we can provide that by using the 'data' keyword argument.
#This would enable generation of plots with the strings corresponding to these variables
```

```python
data = {'a' : np.arange(50),
        'c' : np.random.randint(0,50,50),
        'd' : np.random.randn(50)}

data['b'] = data['a'] + 10 * np.random.rand(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['Group A', 'Group B', 'Group C']
values = [36, 12, 90]

#The figsize attribute allows us to specify the width and height of a figure in unit inches.
plt.figure(figsize=(15, 5))

#The number within the brackets of the subplot denotes the grid of the subplot matrix and which position 
#in that subplot matrix does this specific plot refer to
#131 below mean --> '1' by '3' matrix and the subplot at position '1' of that subplot matrix
plt.subplot(131)
plt.bar(names, values, color='y')

plt.subplot(132)
plt.scatter(names, values, c='g')

plt.subplot(133)
plt.plot(names, values, '--c*')

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
#Controlling Line Properties
#Lines have attributes that can be set. We now explore the various methods by which we can invoke the same
```

```python
#TYPE 1 - USING KEYWORD ARGUMENTS
```

```python
#In this type, we use 'linewidth' as an argument to specify the width of the lines in the plot
data = {'z' : np.arange(50),
        'l' : np.random.randn(50)}
plt.plot('z', 'l', linewidth=2.0, data=data)
```

```python
#TYPE 2 - TUPLE UNPACKING OF THE LINE2D INSTANCE
```

```python
#Here we have explicitly set the anti-aliasing methods to false by invoking the setter method
#Thus there is no anti-alias rendering in the graph below
data = {'z' : np.arange(50),
        'l' : np.random.randn(50)}
line, = plt.plot('z', 'l', data=data)
line.set_antialiased(False) # turn off antialiasing
```

```python
#Let us place the 2 graphs side by side --> one with and one without anti-aliasing
```

```python
data = {'z' : np.arange(50),
        'l' : np.random.randn(50)}

plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.plot('z', 'l', data=data)

plt.subplot(122)
line, = plt.plot('z', 'l', data=data)
line.set_antialiased(False) # turn off antialiasing

plt.suptitle('Comparision of Anti-aliasing effect')
plt.show()
```

```python
#TYPE 3 - USING THE SETP() COMMAND
```

```python
#setp basically offers the flexibility to set one or multiple parameters together by passing them all in one function

data = {'z' : np.arange(50),
        'l' : np.random.randn(50)}

lines = plt.plot('z', 'l', data=data)
# use keyword args to set the parameters as desired
plt.setp(lines, color='m', linewidth=1.0)
# or MATLAB style string value pairs
plt.setp(lines, 'color', 'm', 'linewidth', 1.0)

#Note: Easier to understand the python syntax as it does not use the value pairs concepts
```

```python
#One can also additionally load all the settable line properties
#This can be done by calling the setp() function with a line as argument

data = {'z' : np.arange(50),
        'l' : np.random.randn(50)}

lines = plt.plot('z', 'l', data=data)

plt.setp(lines)

#The settable line properties displayed as output can be set as part of the setp() function for formatting purposes
```

```python
#MATLAB and pyplot use the current figure and current axes to apply plotting commands
#gca() returns current axes and gcf() returns the current figure
#In order to understand these better, let us use an example and create 2 subplots for the same


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()


#All explanations for subplot matrices etc. have been defined in previous graphs
#You may also alter the axes in subplots to perform various functions such as 
#Sharing x or y axes across graphs, per row, per column etc.
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])

plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

#These functions are called again below to set the active figure and then the active subplot
#Then you can edit the properties for it as required
plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title

plt.figure(2)
plt.title('Bigger version of 4,5,6')
```

```python
#Working with text
#Here we will try to add text in any arbitrary location using the text() command
#The xlabel(), ylabel() and title() are used to add text in the indicated locations

#The advantage of using text() commands is that, because it embeds fonts directly in output documents,
#e.g., for postscript or PDF, what you see on the screen is what you get in the hardcopy.
#Additionally, Matplotlib implements a large number of TeX math symbols and commands, 
#supporting mathematical expressions anywhere in your figure.
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=2, facecolor='c', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(120, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
#Similar to formatting line properties, one can also use the same concept to format the text parameters as well
#This is achieved again by using the setp() command, or specifying them in the text formatting directly

#We show this and more text properties by modifying the above histogram texts

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=2, facecolor='r', alpha=0.75)


t = plt.xlabel('Smarts', fontsize=14, color='red', fontweight='bold')
u = plt.ylabel('Probability', backgroundcolor='y', color='k', fontstyle='italic')
plt.title('Histogram of IQ')
plt.text(120, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
#Using mathematical expressions in text

#matplotlib accepts TeX equation in any text expression
#The syntax for that is to enclose the TeX expression as --> '$<your expression here>$'

#Let us modify the above histogram graph to show a few more TeX Expressions


mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=2, facecolor='m', alpha=0.75)


t = plt.xlabel('Smarts - 'r'$\Psi$', fontsize=14, color='red', fontweight='bold')
u = plt.ylabel('Probability - 'r'$\rho$', backgroundcolor='y', color='k', fontstyle='italic')
plt.title('Histogram of IQ - 'r'$\mathcal{IQ}$')
plt.text(120, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
#Annotating Text

#A common use for text is to annotate some feature of the plot, and the annotate() method provides helper functionality to make annotations easy. 
#In an annotation, there are two points to consider: the location being annotated represented by the argument xy and the location of the text xytext. 
#Both of these arguments are (x,y) tuples.

#Creating a subplot 
ax = plt.subplot(111)

#Specifying the range of the x axes
t = np.arange(0.0, 5.0, 0.01)

#Specifying the function of the plot
s = np.cos(2*np.pi*t)

line, = plt.plot(t, s, '--c',lw=2)

#Sample of Annotation below. The point is 2,1 and the text will be displayed at 3,1.5
plt.annotate('Local Maximum', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05, color='m'),
             backgroundcolor='m', color='k'
             )

#Another Example of Annotation
plt.annotate('Local Minimum', xy=(2.5, -1), xytext=(1.5, -1.5),
             arrowprops=dict(facecolor='black', shrink=0.05, color='y'),
             backgroundcolor='y', color='k',
             )


plt.ylim(-2, 2)
plt.show()
```

```python
#Logarithmic and other nonlinear axes

#We've been working with linear scales till now, but it is possible to change the scale to a non linear axis as well
#This can be done by using the xscale() or yscale() function etc.

from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
