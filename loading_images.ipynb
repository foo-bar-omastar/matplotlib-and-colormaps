```python
img = imageio.imread('C:/Users/savya/OneDrive/Desktop/Scientific Visualization/Exercise 2/TEST_IMG.jpg')

print('Type of the image : ' , type(img))
print()
print('Image size {}'.format(img.size))
print('Shape of the image : {}'.format(img.shape))
print('Image Height {}'.format(img.shape[0]))
print('Image Width {}'.format(img.shape[1]))
print('Dimension of Image {}'.format(img.ndim))
```

```python
'''
Let's pick a specific pixel located at 2500 th Rows and 2500 th Column. 
And view the RGB value gradually. 
'''
img[ 2500, 2500 ]

# A specific pixel located at Row : 100 ; Column : 50 
# Each channel's value of it, gradually R , G , B
print('Value of only R channel {}'.format(img[ 2500, 2500, 0]))
print('Value of only G channel {}'.format(img[ 2500, 2500, 1]))
print('Value of only B channel {}'.format(img[ 2500, 2500, 2]))
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.title('R channel')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))

plt.imshow(img[ : , : , 0])

plt.subplot(132)
plt.title('G channel')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))

plt.imshow(img[ : , : , 1])

plt.subplot(133)
plt.title('B channel')
plt.ylabel('Height {}'.format(img.shape[0]))
plt.xlabel('Width {}'.format(img.shape[1]))

plt.imshow(img[ : , : , 2])

plt.suptitle('R-G-B Channel Plotting')
plt.show()
```
