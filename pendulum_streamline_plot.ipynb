The velocity of the pendulum is the highest in the center of the graph, along the y-axis = 0 line. 
This is because the vertical downward acceleration is maximum when the pendulum is at its origin/equilibrium position.
Additionally,there is a maximum downward acting gravitational force, which is not split into horizontal and vertical components as would have been the case when the pendulum were going upwards or coming downwards.

As we can see, some curves are circular and some are not, representing the motion of the pendulum. A circular motion represents a complete oscillation of the pendulum. 
The size of the circle represents the magnitude of the initial angle of displacement for the oscillation of the pendulum.
The lines that are not circular represent a full circular swing of the pendulum / incomplete oscillations.
